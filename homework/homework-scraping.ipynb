# PART 2: Scraping Le Monde

Scrape the content of https://www.lemonde.fr/Links to an external site. and save it as a CSV.

We want: **titles, subhead, article URL, whether it's premium or not, byline, article type, image URL.**

Bonus, if you want to get fancy:

> Make the CSV file auto-updating. Use this tutorial (videoLinks to an external site., textLinks to an external site.) but just ignore the visualization/datawrapper aspect

Tips:

* You're going to be missing some article URLs - you should use try/except on them to avoid them being a problem, buuuut after that I would recommend printing the element to see why your approach isn't working (it's 100% possible to get all of the URLs!)
* If you want to make sure the element you're getting has a specific attribute, you can do something like .select_one("a[href]"). That will only give you anchor tags (links) that have hrefs.
* Instead of yes/no for the premium question, you can think of it as "put text in this column if the article is premium, don't put text in it if it is not"

```python
import requests
from bs4 import BeautifulSoup

lemonde_url = 'https://www.lemonde.fr/en/'

response = requests.get(lemonde_url)
lemonde = BeautifulSoup(response.text)
```

```python
# NEED: titles, subhead, article URL, whether it's premium or not, byline, article type, image URL.

# name of class = article article__title
items = lemonde.find_all('a', class_='article')

len(items)
```

```python
# # Check items

# for item in items:
#     print("-----")
#     print(item)
```

```python
# Check

for item in items:
    print("-----")

    # Headline
    print(item.select_one('.article__title, .article__title-label').text)

    # Subhead
    try:
        print(item.find(class_='article__desc').text)
    except:
        print('No subhead on frontpage')

    # Article URL
    print(item.get('href', 'No URL'))
    
    # Subscription
    try:
        print(item.find(class_='sr-only').text)
    except:
        print('Free')
    
    # Byline
    try:
        print(item.find(class_='article__byline').text)
    except:
        print('No byline on frontpage')
    
    # Article type
    try:
        print(item.find(class_='article__kicker').text)
    except:
        print('No article type on frontpage')

    # Image
    try:
        print(item.find('img').get('data-lazy','No image on frontpage'))
    except:
        print('No image on frontpage')
```

```python
# Save into list of dictionaries
rows = []

for item in items:
    row = {}

    # Headline
    try:
        row['headline'] = item.select_one('.article__title, .article__title-label').text
    except:
        row['headline'] = None

    # Subhead
    try:
        row['subhead'] = item.find(class_='article__desc').text
    except:
        pass

    # Article URL
    row['url'] = item.get('href', None)
    
    # Subscription
    try:
        row['subscription'] = item.find(class_='sr-only').text
    except:
        row['subscription'] = 'Free'
    
    # Byline
    try:
        row['byline'] = item.find(class_='article__byline').text
    except:
        row['byline'] = None
    
    # Article type
    try:
        row['article_type'] = item.find(class_='article__kicker').text
    except:
        pass

    # Image
    try:
        row['img_url'] = item.find('img').get('data-lazy',None)
    except:
        row['img_url'] = None

    rows.append(row)

len(rows)
```

```python
import pandas as pd

lemonde_df = pd.json_normalize(rows)

lemonde_df.head()
```

```python
lemonde_df.to_csv('lemonde-frontpage-articles.csv', index=False)
```

# Part Four: Bankruptcy Court Western District of Tennessee

Visit https://www.tnwb.uscourts.gov/Search/Search.aspxLinks to an external site. and search for "CAR." Scrape the results into a CSV, with four columns: the URL to the case, the name of the case, the category (e.g. "Judge's Opinions), the additional details (terms match/size/pdf URL).

Bonuses, if you want to get fancy:

Split up the additional details into multiple columns
Download all of the PDFs of the cases

**Tips:**

* There are only 10 results, and so many pages! ...maybe there's a secret way to get them all on one page?
* Check the class you're using and see if it matches the number of results (it probably doesn't!). Inspect the page to find out why. You have two options: use something like we did in class with item.select("h1, h2") – but slightly different, since we're talking about classes – or have two separate loops.
* .split is often a convenient way to separate semi-structured text
* Downloading PDFs in Python probably does not involve wget (unless you really want to)

```python
import requests
from bs4 import BeautifulSoup

court_url = 'https://www.tnwb.uscourts.gov/Search/search.aspx?zoom_query=car&zoom_page='

court_files = []

for page in range(1,15):
    response = requests.get(court_url + str(page))
    court_records = BeautifulSoup(response.text, 'html.parser')

    results = court_records.select('.result_block, .result_altblock')

    for result in results:
        court_file = {}
        
        # Case URL
        try:
            court_file['case_url'] = result.find('a').get('href', 'No URL')
        except:
            pass
    
        # Case Name
        court_file['case_name'] = result.find('a').get_text()
        
        # Category
        court_file['category'] = result.find(class_='category').get_text(strip=True).strip('[]')
    
        # Context
        court_file['description'] = result.find(class_='context').get_text(strip=True)
        
        # Additional Details:
        other_details = result.find(class_='infoline').get_text(strip=True).split('-')

        court_file['term_match'] = other_details[0].strip('Terms Matched: ') if len(other_details) > 0 else ''
        court_file['size'] = other_details[1].strip() if len(other_details) > 1 else ''
        court_file['pdf_url'] = other_details[2].replace('URL: ', '').strip() if len(other_details) > 2 else ''

        # Append
        court_files.append(court_file)
```

```python
import pandas as pd

court_df = pd.DataFrame(court_files)

court_df.head()
```

```python
court_df.to_csv('bankruptcy_court_files.csv', index=False)
```
